```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sqlalchemy import create_engine

engine = create_engine("mysql+pymysql://root:gearup4me@localhost:3306/greenspace_analysis")
df_articles = pd.read_sql("SELECT * FROM articles",engine)
df_criteria = pd.read_sql("SELECT * FROM criteria",engine)
df_data_sources = pd.read_sql("SELECT * FROM data_sources",engine)
df_findings = pd.read_sql("SELECT * FROM findings",engine)
df_geo_tools = pd.read_sql("SELECT * FROM geo_tools",engine)
df_themes = pd.read_sql("SELECT * FROM themes",engine)

```

<h1>Top 15 Geo Tools<h1>

```python
df_geo_tools_top_15 = df_geo_tools["tool"].value_counts().iloc[:15].index
geo_tools_graph1 = sns.countplot(data = df_geo_tools, y="tool", order=df_geo_tools_top_15, hue="tool")
geo_tools_graph1.set_xlabel("Count", fontsize =11)
geo_tools_graph1.set_ylabel("Geospatial Tool", fontsize =11)
geo_tools_graph1.set_title("Most Common Tools Used")
```

```python


# df_criteria_top_5 = df_criteria["criterion"].value_counts().iloc[:7].index
criteria_graph1 = sns.countplot(data = df_criteria, y="category", order = df_criteria["category"].value_counts().index, hue="category")
criteria_graph1.set_xlabel("Count", fontsize =11)
criteria_graph1.set_ylabel("Criterion Category", fontsize =11)
criteria_graph1.set_title("Most Common Criteria Used")
```
